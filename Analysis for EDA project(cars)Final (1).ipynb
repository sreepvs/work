# Data Cleaning

### changing columns data types

```python
#defing a  Function to clean and convert the data type(obj to numeric) of distance_kms
def convert_to_number(value):
    # Remove non-numeric characters
    cleaned_value = ''.join(char for char in value if char.isdigit() or char == '.')
    
    # Convert to integer or float
    return int(cleaned_value) if cleaned_value.isdigit() else float(cleaned_value)

# Applying  the function to the DataFrame
Data_Frame['Distance_kms_driven'] = Data_Frame['Distance_kms_driven'].apply(convert_to_number)
```

```python
# Calculate the price in lakhs and store it as a numeric column (float)
Data_Frame["Price(lakhs)"] = Data_Frame["Price(₹)"] / 100000

# Round to 1 decimal place (if desired)
Data_Frame["Price(lakhs)"] = Data_Frame["Price(lakhs)"].round(1)
### droping privious price column 
Data_Frame.drop('Price(₹)',axis=1,inplace=True)
```

## 1.How can we suggest cars based on user preferences such as brand, fuel type, and year of manufacture?

```python
import pandas as pd

# Sample Data (Replace this with your actual Data_Frame)

# Creating DataFrame from the data
Data_Frame = pd.DataFrame(Data_Frame)

# Function to filter cars based on user inputs
def filter_cars(brand_name=None, year_range=None, fuel_type=None):
    filtered_df = Data_Frame.copy()

    # Filter by brand name if provided
    if brand_name:
        filtered_df = filtered_df[filtered_df['Brand'].str.contains(brand_name, case=False, na=False)]
    
    # Filter by year range if provided (year_range is a tuple: (start_year, end_year))
    if year_range:
        start_year, end_year = year_range
        filtered_df = filtered_df[
            (filtered_df['Year of Manufacture'] >= start_year) & 
            (filtered_df['Year of Manufacture'] <= end_year)
        ]
    
    # Filter by fuel type if provided
    if fuel_type:
        filtered_df = filtered_df[filtered_df['Fuel Type'].str.contains(fuel_type, case=False, na=False)]

    # Reset index after filtering to avoid repeated indices
    filtered_df = filtered_df.reset_index(drop=True)

    return filtered_df

# Main block to take user inputs and display filtered results
def main():
    print("Welcome! Let's filter the cars based on your preferences.")

    print("****************************************************************************************************************************************")

    # Input for car brand
    brand_input = input("Enter car brand name: ").strip()

    # Input for year of manufacture range
    start_year_input = int(input("Enter start year of manufacture: ").strip())
    end_year_input = int(input("Enter end year of manufacture: ").strip())

    # Input for fuel type
    fuel_input = input("Enter fuel type (e.g., Petrol, Diesel, Hybrid): ").strip()
    print("****************************************************************************************************************************************")

    # Filter the cars based on user input
    result = filter_cars(
        brand_name=brand_input,
        year_range=(start_year_input, end_year_input),
        fuel_type=fuel_input
    )

    # Display the filtered results as a DataFrame
    if not result.empty:
        print(f"\nMatching Cars (Total: {len(result)} cars found):")
        print(result.to_string(index=False))  # This will print the DataFrame without showing the row index
    else:
        print("No matching cars found.")

# Run the main function
if __name__ == "__main__":
    main()
```
