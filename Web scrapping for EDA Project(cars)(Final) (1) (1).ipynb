```python
final=[] 
prices=[]
ki=[]
for i in range(2,32):
    url = 'https://www.cartrade.com/buy-used-cars/hyderabad/c/page-'+str(i)+'/#so=-1&sc=-1&city=2'
    #print(url)
    page=requests.get(url)
    soup=BeautifulSoup(page.content)
    
    for j in soup.find_all('a',class_='font-muli'): 
        final.append(" ".join(re.findall('\w+',j.text)))
    for g in soup.find_all('div',class_='price_EMI font-muli'): 
        prices.append(g.text)
    for i in soup.find_all('span',class_='carInfo'): 
        ki.append(i.text.strip())
```

```python
### extracting price by removing unwanted things like \n,\t
price_pattern = re.compile(r'₹[\d,]+')
clean_prices = [re.search(price_pattern, price.strip()).group(0) for price in prices if re.search(price_pattern, price.strip())]
```

```python
clean_prices      # cleared data of the prices of the cars
```

```python
len(clean_prices)
```

```python
## extracting kms,location,fuel type by removing useless things(\n,\t..)
kmfuelloc=[]
for i in ki:
    clean_text = re.sub(r'[\n\t\xa0]', '', i)
    kmfuelloc.append(clean_text)
```

```python
# creating a data frame by using name,price,km)
df = pd.DataFrame({'Name': final,
    'Price': clean_prices,
    'Km': kmfuelloc})
```

```python
df
```

```python
## from **Name** column extracting the year,description of the cars
## by using a user defined function
def split_text(text):
    match = re.match(r'(\d{4})\s+(.*)', text)
    if match:
        return match.groups()
    return (None, text)
```

```python
## applying function to the data frame by using apply() function
df[['Year of Manufacture', 'Description']] = df['Name'].apply(split_text).apply(pd.Series)
```

```python
df.head()
```

```python
# finding List of car brands by using span tag
Brands=[]
for i in soup.find_all('span', class_ = 'pull-left'):
    Brands.append(i.text.strip())
Brands.append('Mercedes Benz')




brand_pattern = '|'.join(Brands)


#applying  function to extract brand and rest of the string
def extract_brand_and_rest(text):
    match = re.match(rf'({brand_pattern})\s+(.*)', text)
    if match:
        return match.groups()
    else:
        return (None, text)
```

```python
## spliting description column into two columns(brand&model)
df[['Brand', 'Model']] = df['Description'].apply(extract_brand_and_rest).apply(pd.Series)
```

```python
df.head()
```

```python
## spliting km column into 3 columns(Distance_kms_driven,fuel,location) & dropping km column
df[['Distance_kms_driven', 'Fuel Type', 'Location']] = df['Km'].str.split('|', expand=True)
df.drop('Km',axis=1,inplace=True)
df.head()
```

```python
## spliting location column into two columns(loc,useless)
df[['Location','useless']] = df['Location'].str.split(', ', expand=True)
df.head()
```

```python
## removing useless col
df.drop('useless',axis=1,inplace=True)
```

```python
df.head()
```

```python
## converting dtype of year from object to numeric
df['Year of Manufacture'] = df['Year of Manufacture'].astype(int)
```

```python
# by defing user defined function for creating new column based on  year column
def car_age_category(row):
    if row['Year of Manufacture'] > 2021:
        return 'New'
    elif row['Year of Manufacture'] > 2016:
        return 'Moderate'
    else:
        return 'Old'
        
df['Grade of the Vehicle'] = df.apply(car_age_category, axis=1)
```

```python
df.drop('Name',axis=1,inplace=True)
df.drop('Description',axis=1,inplace=True)
```

```python
# converting dtype of price from obj(str) to numeric(float)
df['Price(₹)'] = df['Price'].replace({'₹': '', ',': ''}, regex=True).astype(int)
```

```python
df.drop('Price',axis=1,inplace=True)
```
